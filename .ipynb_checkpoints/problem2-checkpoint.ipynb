```python
import networkx as nx
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
G = nx.read_gml("SI301_social_network.gml")
```

```python
selfLoops = G.selfloop_edges()
selfLoops = G.remove_edges_from(selfLoops)
for node, degree in G.degree().items():
    if degree == 0:
        G.remove_node(node)
        
undirected = G.to_undirected(reciprocal = True)

nx.draw(G, with_labels=True)
```

```python
print(nx.number_of_nodes(G))
```

```python
#function to reset every behavior to A and every node to unseen

def behaviorSet(graph):
    nx.set_node_attributes(graph, 'behavior', 'A')
    nx.set_node_attributes(graph, 'seen', 'f')

```

```python
def getTotalChange(graph):
    counter = 0
    for node in graph.nodes(data= True ):
        if node[1]['behavior'] == 'B':
            counter += 1
            
    return counter 
```

```python
def behaviorCascade(graph, threshold):
    solutionNodes = []
    solution = []
    for node in graph.nodes(data= True ):
        behaviorSet(graph)
        searchNode = list(node)
        numConverted = 0
        searchNode[1]['behavior'] = 'B'
        searchNode[1]['seen'] = 't'
        currentSearch = []
        currentSearch.append(searchNode)
        
        while not len(currentSearch) == 0:
            currNode = currentSearch.pop()
            totalA = 0
            totalB = 0
            currNode = tuple(currNode)
            graph.node[currNode[0]]['seen'] = 't'
            for neighbor in graph.neighbors(currNode[0]):  
                node=graph.node[neighbor]
                if node['seen'] == 'f':
                    currentSearch.append((neighbor, node))
                    #node['seen'] == 't'
                if node['behavior'] == 'B':
                    totalB += 1
                else:
                    totalA += 1
                
            if not totalB == 0:
                if not totalA == 0:
                    if((totalB / float(totalA + totalB)) >= threshold):
                        currNode[1]["behavior"] = 'B'
                        for neighbors in graph.neighbors(currNode[0]):
                            recheckNode=graph.node[neighbors]
                            if recheckNode['seen'] == 't':
                                if recheckNode['behavior'] == 'A':
                                    recheckNode['seen'] = 'f'
                                    currentSearch.append((neighbor, recheckNode))
                                   
        
        numConverted = getTotalChange(graph)
        #print(node)
        solution.append([numConverted, searchNode[0]])
        
    
    print("solution list")
    print((solution))
    print("max")
    print(max(solution))
              
```

```python
behaviorCascade(G, .3)
```
